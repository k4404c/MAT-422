#### **Probability Axioms**:
- The sample space of an experiment, denoted by S, is the set
of all possible outcomes of that experiment.
- An event is any collection (subset) of outcomes contained in
the sample space S
 * Simple - consists exaclty one outcome
 * Compound - consists more than one outcome
- Given an experiment and a sample space S, the probability
distribution is a function which assign to each event A a number P(A), called
the probability of the event A, which will give a precise measure of the chance
that A will occur. Probability assignments must satisfy the following:
 * For any event A, 1 ≥ P(A) ≥ 0
 * P(S) = 1
 * if $A_1, A_2, A_3, . . .$ is an infinite collection of disjoint events, then:
$P (A_1 ∪ A_2 ∪ A_3 ∪ · · · ) =∑_{i=1}^∞P (A_i)$
 * For any event A, P(A) + P (A′) = 1, from which P(A) = 1 − P (A)
 * When events A and B are mutually exclusive, P(A ∪ B) = P(A) + P(B)
 * For any two events A and B, P(A ∪ B) = P(A) + P(B) − P(A ∩ B)
- In a simply and yet common experiment consisting of N out-
comes, it is reasonable to assign equal probabilities to all N simple events.
That is, if there are N equally likely outcomes, the probability for each is 1 / N
 * Now consider an event A, with N(A) denoting the number of outcomes
contained in A and we have
P(A) = $\frac{N(A)}{N}$

#### **Conditional Probability**:
- Conditional probability is the likelihood of an event or outcome
occurring, based on the occurrence of a previous event or outcome.
* For any two events A and B with P(B) > 0, the conditional
probability of A given that B has occurred is defined by:
 * P(A | B) = $\frac{P(A ∩ B)}{P(B)}$
* Two events A and B are independent if P(A | B) = P(A) ( or
P(A ∩ B) = P(A) · P(B)), and are dependent otherwise.
* Events $A1, . . . , A_n$ are mutually independent if for every k(k =
2, 3, . . . , n) and every subset of indices $i_1, i_2, . . . , i_k$,
P $(A_{i1} ∩ A_{i2} ∩ . . . ∩ A_{i}) = P (A_{i1} ) · P (A_{i2} ) · · · · · · P (A_{ik})$

#### **Discrete Random Variables**:
* A discrete random variable is a random variable whose possible values either constitute a finite set or else can be listed in an infinite
sequence.
*PMF(x) → The probability value will be x
*CDF(x) → The probability value is ≤ x
*Poissan Distribution - Describes the probability of a given number of events occurring in
a fixed interval of time or space if these events occur with a known constant
mean rate and independently of the time since the last event.
 * PMF for a poisson distribution has the form, $p(x;) = \frac{e^{−μ} μ^x}{x!}
 , x = 0, 1, 2, 3, . . . $
* Expected value of x, $E(x) = μ_X = ∑_{x∈D}x · p(x)$
* Variance of x, $V (X) = σ^2_X = ∑_D(x − μ)^2 · p(x) = E [(X −μ)2$
 * $σ_X =√σ^2_X$
* If X is a binomial random variable with parameters n, p, then , E(X) = np,V (X) = np(1 − p),
σX = √np(1 − p).\
 If X is a Poisson distribution with parameter
μ, then E(X) =
μ,V (X) =
μ

```python
def PMFpoisson(lam, k):
  return ((lam**k)  * math.exp(lam*-1)) / (math.factorial(k)*1.0)


lam = 1
ylist = []

xi = list(range(0,100))
for i in range(0,11):
  for k in xi:
    ylist.append(PMFpoisson(lam+i,int(k)))
  plt.plot(xi,ylist, label = f"lambda = {lam+i}")
  ylist.clear()
  plt.legend()
plt.show()

```

#### **Continuous Random Variables**:
* A random variable is continuous if both of the following apply
 * Its set of possible values consists all numbers in a single interval on the
number line.
 * P(X = c) = 0 for any possible value individual c.

* $pdf(x) = P(a ≤ X ≤ b) =∫^b_af (x)dx$
* Expected Mean = $μ_X = E(X) =∫^∞_{−∞}x · f (x)dx$
* Variance = $σ^2_X  = V (X) =∫^∞_{−∞}(x −μ)^2 · f (x)dx = E [(X −μ)^2$
* Standard Deviation = $σ_X = √V (X)$



Continuous Distributions:
* Exponential
 * $pdf(X) = f (x;λ ) = λ e^{−λ x}$ if $ x ≥ 0$
 *  $pdf(X)  = f (x;λ ) = 0 $        otherwise
 * $E(X) =∫^∞_0xλ e^{−λx}dx$
 * μ = $\frac{1}{λ}$ , $σ^2 = \frac{1}{λ^2}$

* Normal
 * $pdf(X) = f (x;μ,σ ) = \frac{1}{√2πσ}e^{−(x−μ)^2/(2σ^2)}$ , $− ∞ < x < ∞ $


```python
def PDFexponential(x, lam):
  return lam*math.exp(-1*lam * x)

xi = list(range(0, 100))
lam = 2
ylist = []
for i in range(0,11):
  for k in xi:
    ylist.append(PDFexponential(k/100.0,lam + i))
  plt.plot(xi,ylist, label = f"lambda = {lam + i }")
  ylist.clear()
  plt.legend()
plt.show()
```

```python
def PDFnormal(x, mean, stddev):
  return (1 / (math.sqrt(2*math.pi*stddev)) * math.exp(-1 * (x - mean)**2 / (2 * stddev**2)))



xi = list(range(-500, 500))
mean = 0
stddev = 1
ylist = []
for i in range(0,11):
  for k in xi:
    ylist.append(PDFnormal(k/100.0, mean, stddev+i))
  plt.plot(xi,ylist, label = f"std dev = {stddev+i}")
  ylist.clear()
  plt.legend()
plt.show()

for i in range(0,11):
  for k in xi:
    ylist.append(PDFnormal(k/100.0, mean + i, stddev))
  plt.plot(xi,ylist, label = f"mean = {mean+i}")
  ylist.clear()
  plt.legend()
plt.show()

for i in range(0,11):
  for k in xi:
    ylist.append(PDFnormal(k/100.0, mean + i, stddev + i))
  plt.plot(xi,ylist, label = f"mean = {mean+i}, std dev = {stddev + i}")
  ylist.clear()
  plt.legend()
plt.show()
```
