```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import make_blobs

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import svm
import pandas as pd

import scipy.stats as st

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.svm import SVC
from sklearn.cluster import KMeans

from sklearn.metrics import adjusted_rand_score
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import roc_curve, roc_auc_score


from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

cancer = datasets.load_breast_cancer()

df_cancer = pd.DataFrame(np.c_[cancer['data'],cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
display(df_cancer.head())
print(cancer.target.shape)
display(cancer.data.shape)

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3,random_state=44)
# 70% training and 30% test

svc_model = SVC()
svc_model.fit(X_train, y_train)
y_predict = svc_model.predict(X_test)

print(classification_report(y_test, y_predict))

# Confusion Matrix for the original dataset
cm_original = confusion_matrix(y_test, y_predict)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_original, annot=True, fmt='g', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.title('Confusion Matrix - Original Dataset')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curve for SVM on the original dataset
svm_probs_original = svc_model.decision_function(X_test)
fpr_svm_original, tpr_svm_original, _ = roc_curve(y_test, svm_probs_original)
roc_auc_svm_original = roc_auc_score(y_test, svm_probs_original)

plt.figure(figsize=(8, 6))
plt.plot(fpr_svm_original, tpr_svm_original, color='blue', label=f'SVM - Original (AUC = {roc_auc_svm_original:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('ROC Curve - SVM on Original Dataset')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

selected_features = []

correlation_threshold = 0.7

for i in range(cancer.data.shape[1]):
    feature_data = cancer.data[:, i]
    feature_name = cancer.feature_names[i]
    correlation = st.pointbiserialr(cancer.target, feature_data).correlation
    if abs(correlation) >= correlation_threshold:
        selected_features.append(i)

# new dataset with the selected features
selected_data = cancer.data[:, selected_features]

print("Selected Features:", selected_features)
print("Shape of Selected Data:", selected_data.shape)

X_min_train, X_min_test, y_new_train, y_new_test = train_test_split(selected_data, cancer.target, test_size=0.3,random_state=44)


svc_model = SVC()
svc_model.fit(X_min_train, y_new_train)
y_min_predict = svc_model.predict(X_min_test)

print(classification_report(y_new_test, y_min_predict))

# Model Accuracy: how often is the classifier correct?
print("Accuracy original random seed:",metrics.accuracy_score(y_test, y_predict))
print(f"Size of data: {cancer.data.shape[1]}")
print("Accuracy minimized random seed:",metrics.accuracy_score(y_new_test, y_min_predict))
print(f"Size of data: {selected_data.shape[1]}")

# Confusion Matrix for the minimized dataset
cm_minimized = confusion_matrix(y_new_test, y_min_predict)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_minimized, annot=True, fmt='g', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.title('Confusion Matrix - Minimized Dataset')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()


# ROC Curve for SVM on the minimized dataset
svm_probs_minimized = svc_model.decision_function(X_min_test)
fpr_svm_minimized, tpr_svm_minimized, _ = roc_curve(y_new_test, svm_probs_minimized)
roc_auc_svm_minimized = roc_auc_score(y_new_test, svm_probs_minimized)

plt.figure(figsize=(8, 6))
plt.plot(fpr_svm_minimized, tpr_svm_minimized, color='orange', label=f'SVM - Minimized (AUC = {roc_auc_svm_minimized:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('ROC Curve - SVM on Minimized Dataset')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()



# K-means clustering original dataset
from sklearn.metrics import silhouette_score

def plot_silhouette(X, cluster_labels, ax):
    silhouette_avg = silhouette_score(X, cluster_labels)
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10

    ax.set_title("Silhouette plot for K-Means Clustering")
    ax.set_xlabel("Silhouette coefficient values")
    ax.set_ylabel("Cluster label")

# Perform k-means clustering on the original dataset
#kmeans_labels_original = kmeans_original.fit_predict(X_train)
n_clusters = 2

kmeans_original = KMeans(n_clusters=2, random_state=44)
kmeans_original.fit(cancer.data)
original_labels = kmeans_original.labels_
kmeans_labels_original = kmeans_original.fit_predict(X_train)

silhouette_original = silhouette_score(cancer.data, original_labels)

print(f"Silhouette Score for Original Data: {silhouette_original}")
print(f"Size of data: {cancer.data.shape[1]}")
#silhouette Measures how well-separated clusters are. range -1 to 1, 0.5 being moderate


# Silhouette Plot for K-Means on the original dataset
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(1, 2, 1)
ax1.set_xlim([-0.1, 1])
ax1.set_ylim([0, len(X_train) + (n_clusters + 1) * 10])
plot_silhouette(X_train, kmeans_labels_original, ax1)

# Perform k-means clustering on the minimized dataset

#kmeans_labels_minimized = kmeans_minimized.fit_predict(X_min_train)

kmeans_minimized = KMeans(n_clusters=2, random_state=44)
kmeans_minimized.fit(selected_data)
minimized_labels = kmeans_minimized.labels_

kmeans_labels_minimized = kmeans_minimized.fit_predict(X_min_train)
silhouette_minimized = silhouette_score(selected_data, minimized_labels)

print(f"Silhouette Score for Minimized Data: {silhouette_minimized}")
print(f"Size of data: {selected_data.shape[1]}")

# Silhouette Plot for K-Means on the minimized dataset
ax2 = plt.subplot(1, 2, 2)
ax2.set_xlim([-0.1, 1])
ax2.set_ylim([0, len(X_min_train) + (n_clusters + 1) * 10])
plot_silhouette(X_min_train, kmeans_labels_minimized, ax2)

plt.show()

# Comparison of Clustering Results
# Visualize k-means clustering results on both original and minimized datasets
# Scatter plot for the original dataset with feature names

feature_index_1 = 0
feature_index_2 = 1

# Scatter plot for the original
plt.figure(figsize=(8, 6))
scatter_original = plt.scatter(X_train[:, feature_index_1], X_train[:, feature_index_2], c=kmeans_labels_original, cmap='viridis')

plt.title('K-Means Clustering - Original Dataset')
plt.xlabel(cancer.feature_names[feature_index_1])
plt.ylabel(cancer.feature_names[feature_index_2])
plt.colorbar(scatter_original, label='Cluster Labels')
plt.show()

# Scatter plot for the minimized
plt.figure(figsize=(8, 6))
scatter_minimized = plt.scatter(X_min_train[:, feature_index_1], X_min_train[:, feature_index_2], c=kmeans_labels_minimized, cmap='viridis')

# Assuming feature_names is a list containing the names of all features
plt.title('K-Means Clustering - Minimized Dataset')
plt.xlabel(cancer.feature_names[selected_features[feature_index_1]])
plt.ylabel(cancer.feature_names[selected_features[feature_index_2]])
plt.colorbar(scatter_minimized, label='Cluster Labels')
plt.show()


y_true = cancer.target

# Adjusted Rand Index original dataset
ari_original = adjusted_rand_score(y_true, original_labels)
print(f"Adjusted Rand Index for Original Data: {ari_original}")
#ARI Measures the similarity between true labels and predicted clusters.. range -1 to 1, 0.5 being moderate
print(f"Size of data: {cancer.data.shape[1]}")


# Adjusted Rand Index minimized dataset
ari_minimized = adjusted_rand_score(y_true, minimized_labels)
print(f"Adjusted Rand Index for Minimized Data: {ari_minimized}")
print(f"Size of data: {selected_data.shape[1]}")

# Logistic Regression original
logreg_original = LogisticRegression(random_state=44)
logreg_original.fit(X_train, y_train)
y_logreg_original = logreg_original.predict(X_test)

cm_logreg_original = confusion_matrix(y_test, y_logreg_original)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_logreg_original, annot=True, fmt='g', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.title('Confusion Matrix - Logistic Regression (Original Dataset)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Logistic Regression  minimized
logreg_minimized = LogisticRegression(random_state=44)
logreg_minimized.fit(X_min_train, y_new_train)
y_logreg_minimized = logreg_minimized.predict(X_min_test)

cm_logreg_minimized = confusion_matrix(y_new_test, y_logreg_minimized)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_logreg_minimized, annot=True, fmt='g', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.title('Confusion Matrix - Logistic Regression (Minimized Dataset)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# classification reports for logistic regression
print("Classification Report for Logistic Regression on Original Data:")
print(classification_report(y_test, y_logreg_original))
print(f"Size of data: {cancer.data.shape[1]}")


print("\nClassification Report for Logistic Regression on Minimized Data:")
print(classification_report(y_new_test, y_logreg_minimized))
print(f"Size of data: {selected_data.shape[1]}")

# ROC Curve for Logistic Regression on the original dataset
logreg_probs_original = logreg_original.predict_proba(X_test)[:, 1]
fpr_logreg_original, tpr_logreg_original, _ = roc_curve(y_test, logreg_probs_original)
roc_auc_logreg_original = roc_auc_score(y_test, logreg_probs_original)

plt.figure(figsize=(8, 6))
plt.plot(fpr_logreg_original, tpr_logreg_original, color='blue', label=f'Logistic Regression - Original (AUC = {roc_auc_logreg_original:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('ROC Curve - Logistic Regression on Original Dataset')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# ROC Curve for Logistic Regression on the minimized dataset
logreg_probs_minimized = logreg_minimized.predict_proba(X_min_test)[:, 1]
fpr_logreg_minimized, tpr_logreg_minimized, _ = roc_curve(y_new_test, logreg_probs_minimized)
roc_auc_logreg_minimized = roc_auc_score(y_new_test, logreg_probs_minimized)

plt.figure(figsize=(8, 6))
plt.plot(fpr_logreg_minimized, tpr_logreg_minimized, color='orange', label=f'Logistic Regression - Minimized (AUC = {roc_auc_logreg_minimized:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('ROC Curve - Logistic Regression on Minimized Dataset')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

metrics_original = classification_report(y_test, y_predict, output_dict=True)
metrics_minimized = classification_report(y_new_test, y_min_predict, output_dict=True)
metrics_logreg_original = classification_report(y_test, y_logreg_original, output_dict=True)
metrics_logreg_minimized = classification_report(y_new_test, y_logreg_minimized, output_dict=True)

# Extract relevant metrics (accuracy, precision, recall, f1-score)
labels = ['Accuracy', 'Precision', 'Recall', 'F1-Score']

original_data = [
    metrics_original['accuracy'],
    metrics_original['weighted avg']['precision'],
    metrics_original['weighted avg']['recall'],
    metrics_original['weighted avg']['f1-score']
]

minimized_data = [
    metrics_minimized['accuracy'],
    metrics_minimized['weighted avg']['precision'],
    metrics_minimized['weighted avg']['recall'],
    metrics_minimized['weighted avg']['f1-score']
]

logreg_original_data = [
    metrics_logreg_original['accuracy'],
    metrics_logreg_original['weighted avg']['precision'],
    metrics_logreg_original['weighted avg']['recall'],
    metrics_logreg_original['weighted avg']['f1-score']
]

logreg_minimized_data = [
    metrics_logreg_minimized['accuracy'],
    metrics_logreg_minimized['weighted avg']['precision'],
    metrics_logreg_minimized['weighted avg']['recall'],
    metrics_logreg_minimized['weighted avg']['f1-score']
]

# Plotting
bar_width = 0.2
index = np.arange(len(labels))

fig, ax = plt.subplots(figsize=(10, 6))

bar1 = ax.bar(index - bar_width, original_data, bar_width, label='SVM - Original', align='center')
bar2 = ax.bar(index, minimized_data, bar_width, label='SVM - Minimized', align='center')
bar3 = ax.bar(index + bar_width, logreg_original_data, bar_width, label='Logistic Regression - Original', align='center')
bar4 = ax.bar(index + 2 * bar_width, logreg_minimized_data, bar_width, label='Logistic Regression - Minimized', align='center')

ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Comparison of Models on Original and Minimized Datasets')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
